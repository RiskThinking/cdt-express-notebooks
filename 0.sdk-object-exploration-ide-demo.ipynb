This notebook will help you quickly explore the VELO SDK and understand how to interact with Python objects in an IDE or notebook.

It covers:
- Using **Tab-completion** to explore attributes/methods
- Using `?` to get docstrings
- Using `??` to inspect inputs, outsputs, and source code (when available)

This works in VS Code, Google Colab, Kaggle, SageMaker, and more.

```python
from IPython.display import Image
import os
os.environ['RISKTHINKING_API_KEY'] = os.environ.get('RTAI_API_KEY')
```

```python
from velo_sdk.api import APIClient

client = APIClient()
```

# Dot auto complete

```python
# Tab-complete the `client` object to see available methods
client # <- Place cursor after `client` and press .
```

```python
Image(filename='images/auto_complete_example.png', width=500, height=500)
```

This is will you all the objects and methods available for the `client` object. <br>
The box displayed will be scrollable if there are too many methods to display. <br>

Important note: In Kagle the box will not automatically display, you'll need to press `Tab` after the `.` to see the box.

The Process can be chained together like this:

```python
# Tab-complete the `client.assets` object to see available methods
client.assets # <- Place cursor after `client.assets` and press .
```

```python
Image(filename='images/asset_dot_example.png', width=500, height=500)
```

# Question mark for docstring inspection

```python
# A question mark will display the docstring for the method
?client.assets.get_asset
```

```python
# Double question mark for source code inspection, lists input and output parameters as well as list the source code
??client.assets.get_asset
```
